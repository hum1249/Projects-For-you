```python
import pymysql
import pandas as pd

# Connect to MySQL
conn = pymysql.connect(
    host="database-1.chk8ye4ga5gt.us-west-1.rds.amazonaws.com",
    port=3306,
    user="girish",
    password="girish99",
    database="titanic"
)
cursor = conn.cursor()

# Load CSV data into DataFrame
data = pd.read_csv(r"C:\Users\RG\OneDrive\Desktop\flexon data sets\day 1 data sets\tested.csv")

# Check if data is loaded properly
print(data.shape)  # Shows number of rows and columns
print(data.head())  # Shows the first few rows of data

# Replace NaN with None (NULL in MySQL)
data = data.where(pd.notnull(data), "Null")

# Ensure numeric columns are properly converted to the correct type
data['Age'] = pd.to_numeric(data['Age'], errors='coerce')
data['Fare'] = pd.to_numeric(data['Fare'], errors='coerce')

# Check if NaN values are correctly replaced by None
print(data.isnull().sum())  # To verify if NaNs were handled

# Prepare the bulk insert SQL query
sql = """
INSERT INTO titanic (
    PassengerId, Survived, Pclass, Name, Sex, Age, SibSp, Parch, Ticket, Fare, Cabin, Embarked
) VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s)
"""

# Check for empty DataFrame before proceeding
if not data.empty:
    # Prepare data for bulk insert
    values = [tuple(row) for row in data.values]
    
    # Debug: Check values to insert
    print(f"Values to insert: {values[:5]}")  # Shows the first 5 rows to be inserted
    
    # Insert the data in bulk
    cursor.executemany(sql, values)
    
    # Commit and close connection
    conn.commit()
    cursor.close()
    conn.close()

    print("Data loaded successfully")
else:
    print("No data to insert.")
```
